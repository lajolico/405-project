```python
import pandas as pd
```

```python
confirmed = pd.read_csv("data/covid_confirmed_usafacts.csv")
confirmed = confirmed.drop(confirmed.iloc[:, 4 : 865], axis=1)
confirmed = confirmed.drop(confirmed.iloc[:, 218 : 234], axis=1)

column_names = confirmed.columns[4:]
column_groups = [column_names[i:i+7] for i in range(0, len(column_names), 7)]

for group in column_groups:
    group_mean = confirmed[group].mean(axis=1).round()
    group_median = confirmed[group].median(axis=1).round()
    group_mode = confirmed[group].mode(axis=1)[0]
    confirmed[f"{group[0]} - {group[-1]} Mean"] = group_mean
    confirmed[f"{group[0]} - {group[-1]} Median"] = group_median
    confirmed[f"{group[0]} - {group[-1]} Mode"] = group_mode

confirmed = confirmed.drop(confirmed.iloc[:, 4 : 218], axis=1)

confirmed
```

```python
deaths = pd.read_csv("data/covid_deaths_usafacts.csv")
deaths = deaths.drop(deaths.iloc[:, 4 : 865], axis=1)
deaths = deaths.drop(deaths.iloc[:, 218 : 234], axis=1)

column_names = deaths.columns[4:]
column_groups = [column_names[i:i+7] for i in range(0, len(column_names), 7)]

for group in column_groups:
    group_mean = deaths[group].mean(axis=1).round()
    group_median = deaths[group].median(axis=1).round()
    group_mode = deaths[group].mode(axis=1)[0]
    deaths[f"{group[0]} - {group[-1]} Mean"] = group_mean
    deaths[f"{group[0]} - {group[-1]} Median"] = group_median
    deaths[f"{group[0]} - {group[-1]} Mode"] = group_mode

deaths = deaths.drop(deaths.iloc[:, 4 : 218], axis=1)

deaths
```

Calculate mean (rounded to integer value) number of new cases/deaths per week and then calculate basic statistics (mean, median, mode) on that data.

Cases:

```python
# Filter only columns that contain the mean value of the weeks
filter_conf_mean = confirmed.drop(confirmed.filter(regex='Median'), axis=1)
filter_conf_mean = filter_conf_mean.drop(filter_conf_mean.filter(regex='Mode'), axis=1)

# Calculate basic statistics (mean, median, mode) on that mean data
mean_tot = filter_conf_mean.iloc[:,4:].mean(axis=1)
filter_conf_mean['Total Mean Mean'] = mean_tot

median_tot = filter_conf_mean.iloc[:,4:].median(axis=1)
filter_conf_mean['Total Mean Median'] = median_tot

mode_tot = filter_conf_mean.iloc[:,4:].mode(axis=1)[0]
filter_conf_mean['Total Mean Mode'] = mode_tot

filter_conf_mean
```

Deaths:

```python
filter_deaths_mean = deaths.drop(confirmed.filter(regex='Median'), axis=1)
filter_deaths_mean = filter_deaths_mean.drop(filter_deaths_mean.filter(regex='Mode'), axis=1)

# Calculate basic statistics (mean, median, mode) on that mean data
mean_tot_death = filter_deaths_mean.iloc[:,4:].mean(axis=1)
filter_deaths_mean['Total Mean Mean'] = mean_tot_death

median_tot_death = filter_deaths_mean.iloc[:,4:].median(axis=1)
filter_deaths_mean['Total Mean Median'] = median_tot_death

mode_tot_death = filter_deaths_mean.iloc[:,4:].mode(axis=1)[0]
filter_deaths_mean['Total Mean Mode'] = mode_tot_death

filter_deaths_mean
```

```python
world_df = pd.read_csv("data/owid-covid-data-cases.csv", usecols=['location','date','total_cases','new_cases','total_deaths','new_deaths','population'])

# Selecting only the countries that have similar population as the US

pop_similar_usa = ['Indonesia','Pakistan','Nigeria']
dates_used = (world_df['date'] >= '2022-06-01') & (world_df['date'] < '2023-01-01')

world_df = world_df[world_df['location'].isin(pop_similar_usa)]
world_df = world_df.loc[dates_used]
world_df
```

```python
pakistan_df = world_df[world_df['location'].str.contains('Pakistan')].reset_index().drop(['index'], axis = 1)

pakistan_df
```

```python
# Pakistan new cases weekly mean - median - mode
temp = {}
for i in range(0, len(pakistan_df.date), 7):
    if (i + 7) > len(pakistan_df.date):
        j = len(pakistan_df.date) - i -1
        temp[f'{pakistan_df.date[i]} - {pakistan_df.date[i + j]} Mean'] = (pakistan_df.new_cases[i:i + j]).mean()
        temp[f'{pakistan_df.date[i]} - {pakistan_df.date[i + j]} Median'] = (pakistan_df.new_cases[i:i + j]).median()
        temp[f'{pakistan_df.date[i]} - {pakistan_df.date[i + j]} Mode'] = (pakistan_df.new_cases[i:i + j]).mode()[0]
    else:
        temp[f'{pakistan_df.date[i]} - {pakistan_df.date[i+7]} Mean'] = (pakistan_df.new_cases[i:i+7]).mean()
        temp[f'{pakistan_df.date[i]} - {pakistan_df.date[i + 7]} Median'] = (pakistan_df.new_cases[i:i + 7]).median()
        temp[f'{pakistan_df.date[i]} - {pakistan_df.date[i + 7]} Mode'] = (pakistan_df.new_cases[i:i + 7]).mode()[0]


pakistan_df_confirmed = pd.DataFrame(temp, index = ['Pakistan'])
pakistan_df_confirmed
```

```python
# Pakistan new deaths weekly mean - median - mode
temp = {}
for i in range(0, len(pakistan_df.date), 7):
    if (i + 7) > len(pakistan_df.date):
        j = len(pakistan_df.date) - i -1
        temp[f'{pakistan_df.date[i]} - {pakistan_df.date[i + j]} Mean'] = (pakistan_df.new_deaths[i:i + j]).mean()
        temp[f'{pakistan_df.date[i]} - {pakistan_df.date[i + j]} Median'] = (pakistan_df.new_deaths[i:i + j]).median()
        temp[f'{pakistan_df.date[i]} - {pakistan_df.date[i + j]} Mode'] = (pakistan_df.new_deaths[i:i + j]).mode()[0]
    else:
        temp[f'{pakistan_df.date[i]} - {pakistan_df.date[i+7]} Mean'] = (pakistan_df.new_deaths[i:i+7]).mean()
        temp[f'{pakistan_df.date[i]} - {pakistan_df.date[i + 7]} Median'] = (pakistan_df.new_deaths[i:i + 7]).median()
        temp[f'{pakistan_df.date[i]} - {pakistan_df.date[i + 7]} Mode'] = (pakistan_df.new_deaths[i:i + 7]).mode()[0]


pakistan_df_deaths = pd.DataFrame(temp, index = ['Pakistan'])
pakistan_df_deaths
```
