```python
import pandas as pd
import numpy as np


# Importing the main Covid dataset that includes cases, death and population
covid_death = pd.read_csv('datasets/covid_comperhensive.csv', usecols=['countyFIPS','County Name', 'State', 'StateFIPS', '2023-01-10_death','2023-01-11_death','2023-01-12_death','2023-01-13_death','2023-01-14_death','2023-01-15_death', '2023-01-16_death'])
covid_cases = pd.read_csv('datasets/covid_confirmed_usafacts copy.csv', usecols=['countyFIPS','County Name', 'State', 'StateFIPS', '2023-01-10','2023-01-11','2023-01-12','2023-01-13','2023-01-14','2023-01-15','2023-01-16'])

# Selecting only the counties from Virginia(VA)
covid_cases_VA = covid_cases.loc[covid_cases['State']=='VA']


# Aggregating the dataset to get the total cases for the whole dataset for the state of VA
covid_cases_VA_total = covid_cases_VA.groupby(['State','StateFIPS'])
covid_cases_VA_total.agg(np.sum)
```

As you can see the cases in VA are stable for the last week of the data.

```python
# For the deaths:
# Selecting only the counties from Virginia(VA)
covid_death_VA = covid_death.loc[covid_death['State']=='VA']

# Aggregating the dataset to get the total deaths for the whole dataset for the state of VA
covid_death_VA_total = covid_death_VA.groupby(['State','StateFIPS'])
covid_death_VA_total.agg(np.sum)
```

As you can see the death cases in VA are stable for the last week of data. This might be due to the fact that Viginia collects data on a weekly basis.

# Enrichment dataset

```python
import os
import datetime

# Preparing file paths to initiate merging
path = 'datasets/enrichment/'
csv_files = os.listdir(path)
csv_files.sort()

# Creating the whole file path for merging
i = 0
for csv in csv_files:
    csv_files[i] = path + csv
    i += 1

# Adding date to the different columns so that they can be distinguished for certain date
i = 10
tod = datetime.date(2023, 1, i)
temp_date = tod.strftime("%Y-%m-%d")

#Mergin the 7 data files to form one enrichment file
enrich_main = pd.read_csv(csv_files[0], usecols=['state','critical_staffing_shortage_today_yes','critical_staffing_shortage_today_no','inpatient_beds','inpatient_beds_used','inpatient_beds_used_covid'])
enrich_main = enrich_main.rename(columns={col : col + '_' + temp_date for col in enrich_main.columns if col not in['state']})

for csv in csv_files[1:]:
    i += 1
    tod = datetime.date(2023, 1, i)
    temp_date = tod.strftime("%Y-%m-%d")
    
    temp = pd.read_csv(csv, usecols=['state','critical_staffing_shortage_today_yes','critical_staffing_shortage_today_no','inpatient_beds','inpatient_beds_used','inpatient_beds_used_covid'])
    temp = temp.rename(columns={col : col + '_' + temp_date for col in temp.columns if col not in['state']})
        
    enrich_main = enrich_main.merge(temp, on="state", how = 'outer')
    
enrich_main
    
```

# Merging the enrichment dataset with the main COVID-19 dataset

```python
enrich_main = enrich_main.rename(columns={'state':'State'})

cv_19_df = pd.read_csv('datasets/covid_comperhensive.csv')

cv_19_df = cv_19_df.merge(enrich_main, on="State", how = 'outer')

cv_19_df
```
