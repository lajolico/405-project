### Merging COVID-19 Datasets

#### Using Pandas

```python
import pandas as pd
```

#### Loading the data sets onto pandas dataframes

```python
covid_cases = pd.read_csv("data/covid_confirmed_usafacts.csv")
covid_cases = covid_cases.drop(covid_cases.iloc[:, 4:-7], axis=1)
covid_deaths = pd.read_csv("data/covid_deaths_usafacts.csv")
covid_deaths = covid_deaths.drop(covid_deaths.iloc[:, 4:-7], axis=1)
covid_pop = pd.read_csv("data/covid_county_population_usafacts.csv")
```

#### Using pandas.merge to merge

First merging confirmed cases and deaths using common columns, like 'countyFIPS', 'County Name', 'State', 'StateFIPS'.

Then merging the resulted dataframe with the population using 'countyFIPS', 'County Name', 'State'.

Suffixes used for both confirmed cases (_confirmed) and deaths (_death) for clarifications

Finally writing the generated dataframe into a .csv file saved into the '/data' directory

```python
covid_comp = pd.merge(pd.merge(covid_cases, covid_deaths, on=['countyFIPS', 'County Name', 'State', 'StateFIPS'], suffixes=('_confirmed', '_death')), covid_pop, on=['State'])
covid_comp.head(5)
```

```python
covid_comp.to_csv('data/covid_comperhensive.csv', index=False)
```
