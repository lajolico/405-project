```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from math import sqrt
from sklearn.svm import SVR
import warnings
from scipy.stats import pearsonr
from scipy.stats import ttest_ind
from scipy import stats
import statsmodels.api as sm
warnings.filterwarnings('ignore')
```

```python
confirmed_cases = pd.read_csv('data/covid_confirmed_usafacts.csv')
#cleaning data for new york
ny_confirmed = confirmed_cases[confirmed_cases['StateFIPS'] == 36]
ny_confirmed = ny_confirmed.drop(ny_confirmed[ny_confirmed['County Name'] == 'Statewide Unallocated'].index)
ny_confirmed
```

We now know the first date of infections in New York from Jan 22, 2020 was on Mar 2, 2020. From here we can create linear and non-linear regression models.

```python
ny_cases_model = ny_confirmed.copy()
ny_cases_model.drop(ny_confirmed.iloc[:, 4:44], axis=1, inplace=True)

#group cases by county to get sums of cases
ny_cases_model = ny_cases_model.groupby('County Name').sum().drop(['countyFIPS', 'StateFIPS'], axis=1)
ny_cases_model
```

```python
def prediction_model(cases):
    
    group_cases_ny = cases.sum()
    group_cases_ny = group_cases_ny.iloc[0:]
    group_cases_ny.index = pd.to_datetime(group_cases_ny.index)

    temp = group_cases_ny.index
    
    start_date = group_cases_ny.index[0]
    days_since_start = [(date - start_date).days for date in group_cases_ny.index]
    days_and_cases = pd.DataFrame({'Days Since Start': days_since_start, 'New Cases': group_cases_ny.values})

    X =[]
    for i in range(len(temp)):
        X.append(i)
    Y= group_cases_ny.values
    X = np.array(X)
    X= X.reshape(-1,1)
    Y = Y.astype('int')
    alpha = 0.05
    X_train, X_test, y_train, y_test = train_test_split(X, Y)
    
    lr = LinearRegression()
    lr.fit(X_train,y_train)
    
    pr = LinearRegression()
    poly = PolynomialFeatures(degree=3)
    X_poly_train = poly.fit_transform(X_train)
    X_poly_test = poly.transform(X_test)
    pr.fit(X_poly_train, y_train)

    y_pred_lr = lr.predict(X_test)
    y_pred_pr = pr.predict(X_poly_test)  

    rmse_lr = sqrt(abs(mean_squared_error(y_test, y_pred_lr)))
    rmse_pr = sqrt(abs(mean_squared_error(y_test, y_pred_pr)))
    
    print('RMSE:')
    print(f'Linear regression RMSE: {rmse_lr}')
    print(f'Polynomial regression RMSE: {rmse_pr}')
    
    #calculate confidence intervals
    residuals_lr = y_train - lr.predict(X_train)
    ci_lr = np.quantile(residuals_lr, 1 - alpha)
    preds_lr = lr.predict(X_test)
    
    confidence_lr = generate_confidence_interval(preds_lr, ci_lr)
    print('\nLinear Confidence Interval:')
    print(confidence_lr)
    
    residuals_pr = y_train - pr.predict(X_poly_train)
    ci_pr = np.quantile(residuals_pr, 1 - alpha)
    preds_pr = pr.predict(X_poly_test)
    
    confidence_pr = generate_confidence_interval(preds_lr, ci_pr)
    print('\nPolynomial Confidence Interval:')
    print(confidence_pr)
    
    #Calculate prediction intervals
    pred_y_lr = lr.predict(X)
    pred_y_pr = pr.predict(poly.fit_transform(X))
    pi_lr = np.sqrt(abs(preds_lr + rmse_lr**2))
    pi_pr = np.sqrt(abs(preds_pr + rmse_pr**2))
    print('\nLinear Prediction Interval:')
    print('['+str(np.min(pi_lr))+' , '+str(np.max(pi_lr))+']')
    print('\nPolynomial Prediction Interval:')
    print('['+str(np.min(pi_pr))+' , '+str(np.max(pi_pr))+']')
    
    plt.figure(figsize=(20,16))
    plt.scatter(days_and_cases['Days Since Start'], days_and_cases['New Cases'], color='blue')
    plt.plot(days_and_cases['Days Since Start'], lr.predict(days_and_cases[['Days Since Start']]), color='red', label='Linear')
    plt.plot(days_and_cases['Days Since Start'], pr.predict(poly.transform(days_and_cases[['Days Since Start']])), color='lime', label='polynomial')
    plt.ticklabel_format(style='plain')
    plt.legend()
    plt.show()
```

```python
def generate_confidence_interval(preds, ci):
    df = pd.DataFrame()
    df['prediction'] = preds
    if ci >= 0:
        df['upper'] = preds + ci
        df['lower'] = preds - ci
    else:
        df['upper'] = preds - ci
        df['lower'] = preds + ci
        
    return df
```

# Hypothesis Test from Previous Stage

In the previous stage of the project, I identified a hypothesis to test when it came to state demographic data and spread of COVID-19. In this hypothesis, I begged stated:

The data above shows us the relative strength and weakness of the correlation between race demographics among all 50 states and the prevalence of COVID-19 confirmed cases in those 50 states. The data above is very interesting to read. Some notable things regarding the data above:

There is a moderate negative correlation between caucasian americans and covid 19 cases. Asserting that the more caucasian identifying men and women there are in a given state, the less covid cases we tend to see.
Although the correlation is weak, there is a positive correlation between covid cases and most other races, those exceptions are Native American tribes, and Vietnamese.
With this information, we can begin to formulate a null and alternative hypothesis between the enrichment data and the new confirmed covid cases between June 01, 2022 and December 31, 2022.

Null hypothesis: The more people who identify as caucasian in a given state, there is no change in confirmed covid 19 cases in that state

Alternative hypothesis 1: The more people who identify as caucasian in a given state, the number confirmed covid 19 cases there are in that state changes significantly

Alternative hypothesis 2: The more people who identify as caucasian in a given state, the less confirmed covid 19 cases there are in that state

Using this data, we can now begin to test these hypothesis to test for any statistical significance in the data.

```python
#Importing and repeating work to create correlational coefficients
enrichment_dataset = pd.read_csv('data/combined_df.csv')

#cleaning up the data
enrichment_new = enrichment_dataset.drop(enrichment_dataset[enrichment_dataset['County Name'] == 'Statewide Unallocated'].index)

state_population = enrichment_new.groupby('State')['population'].sum()
state_demographics = enrichment_new.groupby('State')['White', 'Black or African American', 'American Indian and Alaska Native', 'Cherokee tribal grouping', 'Chippewa tribal grouping', 'Navajo tribal grouping', 'Sioux tribal grouping', 'Asian', 'Asian Indian', 'Chinese', 'Filipino', 'Japanese', 'Korean', 'Vietnamese', 'Other Asian', 'Native Hawaiian and Other Pacific Islander', 'Guamanian or Chamorro', 'Samoan', 'Other Pacific Islander', 'Some other race'].first()
enrichment_cleaned = state_demographics.merge(state_population, on='State')
enrichment_cleaned = enrichment_cleaned.replace('N', 0)
enrichment_cleaned = enrichment_cleaned.applymap(lambda x: float(x.rstrip('%')) if isinstance(x, str) and x.endswith('%') else x)

#normalize enrichment data by population
enrichment_norm = enrichment_cleaned.copy()
race_cols = enrichment_norm.columns[0:20]

for race in race_cols:
    enrichment_norm[race + '_estimation'] = np.floor(enrichment_norm[race] / 100 * enrichment_norm['population']).astype(int)

enrichment_norm.drop(enrichment_norm.iloc[:, 0:20], axis=1, inplace=True)
enrichment_norm.iloc[:, 1:] = enrichment_norm.iloc[:, 1:].div(enrichment_norm['population'], axis=0)

enrichment_norm
```

```python
#bring in covid_confirmed_usafacts and set up for means of new cases
usa_confirmed = pd.read_csv('data/confirmed.csv')

usa_confirmed = usa_confirmed.drop(usa_confirmed[usa_confirmed['County Name'] == 'Statewide Unallocated'].index)

date_cols = usa_confirmed.columns[4:]

agg_dict = {f'{col}': 'sum' if 'Sum' in col else 'mean' if 'Mean' in col else 'median' if 'Median' in col else lambda x: x.mode()[0] for col in date_cols}

usa_confirmed = usa_confirmed.groupby('State').agg(agg_dict)
usa_confirmed = usa_confirmed.filter(regex=r'Mean')

usa_confirmed[usa_confirmed < 0] = 0

usa_pop = enrichment_norm[['population']].copy()
usa_norm = usa_confirmed.merge(usa_pop, on='State')

cols_to_normalize = usa_norm.columns[0:31]

#normalize usa_facts
for col in cols_to_normalize:
    usa_norm[col] = usa_norm[col] / 100000
    
usa_norm.drop(columns = ['population'], inplace=True)

usa_norm
```

```python
#merging the two dataframes to calculate correlation coefficient
merged_df = pd.merge(enrichment_norm, usa_norm, on="State")
race_cols = enrichment_cleaned.columns[0:20]

correlations = []
for race in race_cols:
    x = enrichment_norm[race + '_estimation']
    y = usa_norm.mean(axis=1)
    corr, _ = pearsonr(x, y)
    correlations.append(corr)

for i in range(len(race_cols)):
    print('{}: {}'.format(race_cols[i], correlations[i]))
```

Using this data, I believe we can draw a conclusion on whether the more people who identify as caucasian in a given state, if there is a change in COVID-19 infections in that state. For this portion, we will focus in on the very first value asserted in the above correlation coefficients...

```python
caucasian_correlation = correlations[0]
caucasian_correlation
```

For calculating t-tests and p-values, we will assume a significance level of 0.05. First we must start by collecting the number of samples. 

In this case, the number of samples would be the number of weeks between 06/01/2022 and 12/31/2022. The number of days in between 06/01/2022 and 12/31/2022 would be 214. Because our data is averaged and normalized by 7 day groupings, we can divide this by 7 to get approx. 31 weeks, which will be our sample size for the t-test.

```python
n = len(usa_norm.columns)
degrees_freedom = n - 1
r = caucasian_correlation

#calculate t-value
t = r * ((n - 2) / (1 - r**2))**0.5

#two tailed p-value
p = stats.t.sf(abs(t), degrees_freedom) * 2

print("Two-tailed t-test")
print("t-value:", t)
print("p-value:", p)

p1 = p/2
print("\nOne-tailed t-test")
print("p-value:", p1)
```

## Results

As we can see from the above t-test, the p-value falls below the significance level proposed. In the case of a two tailed t-test, we can reject the null hypothesis.

Much in the same way, we can also reject the null hypothesis for the one tailed t-test, seeing as how it is also less than the significance level.
